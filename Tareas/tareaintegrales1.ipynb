```python
import matplotlib.pyplot as plt
import numpy as np 

from scipy.integrate import quad, dblquad, tplquad
from scipy.special import legendre
```

```python
f1 = lambda x: x**4
f2 = lambda x: 3*x - x**2

r = np.linspace(-4, 4, 2000)

plt.plot(r, f1(r), color ='blue')
plt.plot(r, f2(r), color ='black')

plt.fill_between(r, f2(r), f1(r),
                 facecolor="orange", # The fill color
                 color='blue',       # The outline color
                 alpha=0.2)          # Transparency of the fill

plt.xlim(-4, 4)
plt.ylim(-3, 3)
```

```python
def integrand(x, y):
    return x

x_lower = 0
x_upper = 1.2
y_upper = lambda x: 3*x - x**2
y_lower = lambda x: x**4

val, abserr = dblquad(integrand, x_lower, x_upper, y_lower, y_upper)

print('valor val1 ', val)
abserr
```

```python
#Ejercicio 4

f1_1 = lambda x: -np.sqrt(x)
f1_2 = lambda x: np.sqrt(x)
f2_1 = lambda x: np.sqrt((3-x)/2)
f2_2 = lambda x: -np.sqrt((3-x)/2)

r = np.linspace(-4, 6, 2000)
r2 = np.linspace(-1, 5, 2000)
r3 = np.linspace(-3, -1, 2000)

plt.plot(r, f1_1(r), color ='blue')
plt.plot(r, f1_2(r), color = 'blue')
plt.plot(r, f2_1(r), color ='black')
plt.plot(r, f2_2(r), color ='black')

plt.fill_between(r2, f1_2(r2), f1_1(r2),
                 facecolor="orange", # The fill color
                 color='blue',       # The outline color
                 alpha=0.2)          # Transparency of the fill

plt.fill_between(r3, f2_1(r3), f2_2(r3),
                 facecolor="orange", # The fill color
                 color='blue',       # The outline color
                 alpha=0.2)          # Transparency of the fill

plt.xlim(-5, 5)  # Set limits for x-axis
plt.ylim(-5, 5)  # Set limits for y-axis
```

```python
func = lambda y, x: y**2 - x 

x_lower = lambda  y: y**2
x_upper = lambda y: 3 - 2*y**2
y_lower = 1
y_upper = -1

#y_upper = lambda x : 10-x

val, abserr = dblquad(func, y_lower, y_upper, x_lower, x_upper)

print(val, abserr)
```
